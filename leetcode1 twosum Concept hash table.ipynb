```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        n=len(nums)

        for i in range(n-1):
            for j in range(i+1,n):
                if nums[i]+nums[j]==target:
                    return [i,j]
```

```python
#For the question we can use dictionary as our hash table: (which I did but with a smarter way)
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        dicti={}
        n=len(nums)
        for i in range(n):
            compliment= target-nums[i]
            if compliment not in dicti:
                dicti[nums[i]]=i # kinda like a reverse dict where key is value of list and value is the index
                
            else:
                return [dicti[compliment],i]

        return [] #final return if none of other is satistied. just so we dont get an error
       
```
